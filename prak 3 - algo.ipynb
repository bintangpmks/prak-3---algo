```python
# Fungsi untuk memeriksa jenis segitiga
def jenis_segitiga(a, b, c):
    # Memeriksa apakah bisa membentuk segitiga
    if (a + b > c) and (a + c > b) and (b + c > a):
        if a == b == c:
            return "Segitiga sama sisi"
        elif a == b or b == c or a == c:
            return "Segitiga sama kaki"
        else:
            return "Segitiga sembarang"
    else:
        return "Tidak dapat membentuk segitiga"

# Mendapatkan input dari pengguna
a = float(input("Masukkan panjang sisi a: "))
b = float(input("Masukkan panjang sisi b: "))
c = float(input("Masukkan panjang sisi c: "))

# Menentukan jenis segitiga
result = jenis_segitiga(a, b, c)

# Menampilkan hasil
print(f"\nHasil: {result}")

import cmath  # Menggunakan cmath untuk mendukung akar kompleks

# Fungsi untuk mencari akar persamaan kuadrat
def akar_persamaan_kuadrat(a, b, c):
    # Menghitung determinan
    D = b**2 - 4 * a * c
    print(f"Determinan (D): {D}")

    # Menghitung akar persamaan kuadrat
    if D > 0:
        x1 = (-b + cmath.sqrt(D)) / (2 * a)
        x2 = (-b - cmath.sqrt(D)) / (2 * a)
        return (x1, x2)
    elif D == 0:
        x1 = -b / (2 * a)
        return (x1,)
    else:
        # Akar kompleks
        x1 = (-b + cmath.sqrt(D)) / (2 * a)
        x2 = (-b - cmath.sqrt(D)) / (2 * a)
        return (x1, x2)

# Mendapatkan input dari pengguna
a = float(input("Masukkan koefisien a: "))
b = float(input("Masukkan koefisien b: "))
c = float(input("Masukkan koefisien c: "))

# Mencari akar persamaan kuadrat
akar = akar_persamaan_kuadrat(a, b, c)

# Menampilkan hasil
if len(akar) == 2:
    print(f"Akar-akar persamaan kuadrat: x1 = {akar[0]}, x2 = {akar[1]}")
elif len(akar) == 1:
    print(f"Akar kembar persamaan kuadrat: x = {akar[0]}")
else:
    print(f"Akar kompleks persamaan kuadrat: x1 = {akar[0]}, x2 = {akar[1]}")1
```
